```
train_path = '/content/drive/MyDrive/emotion detection/train/'
test_path = '/content/drive/MyDrive/emotion detection/test/'
train_csv = 'train_df.csv'
test_csv = 'test_df.csv'
```

# Preprocess gold, xml files

```
from bs4 import BeautifulSoup
```

```
train_df = pd.DataFrame(columns=['text'] + list(range(6)))
with open(train_path + "text.xml") as f:
    soup = BeautifulSoup(f, 'xml')
    for item in soup.find_all('instance'):
      train_df = train_df.append(pd.DataFrame([item.text], index=[int(item['id'])], columns=['text']))
with open(train_path+'emotions.gold') as f:
  for id in train_df.index:
    ems = [int(i) for i in f.readline().split()[1:]]
    for i in range(6):  
      train_df[i].loc[id] = ems[i]
train_df.head()
```

```
train_df.reset_index().to_csv(train_path + train_csv, index=False)
```

```
test_df = pd.DataFrame(columns=['text'] + list(range(6)))
with open(test_path + "text.xml") as f:
    soup = BeautifulSoup(f, 'xml')
    for item in soup.find_all('instance'):
      test_df = test_df.append(pd.DataFrame([item.text], index=[int(item['id'])], columns=['text']))
with open(test_path+'emotions.gold') as f:
  for id in test_df.index:
    ems = [int(i) for i in f.readline().split()[1:]]
    for i in range(6):  
      test_df[i].loc[id] = ems[i]
test_df.head()
```

```
test_df.reset_index().to_csv(test_path + test_csv, index=False)
```
